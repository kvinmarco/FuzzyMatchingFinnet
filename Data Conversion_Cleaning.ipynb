```python
df_a_clean = df_a[["TRANS_DATE", "BILL_NUMBER", "REFF_NUMBER", "TRANS_AMOUNT"]]
df_a_clean = df_a_clean.dropna()
df_a_clean.describe()
```

```python
df_a_clean.head()
```

```python
# add 62 infront of bill number, make it an int64
df_a_clean["BILL_NUMBER"] = "62" + df_a_clean["BILL_NUMBER"].astype(str)
df_a_clean["BILL_NUMBER"] = df_a_clean["BILL_NUMBER"].astype("int64")

df_a_clean.info()
```

```python
df_a_clean
```

```python
df_b.describe()
```

```python
df_b_clean = df_b[["date_time", "bill_number", "reff_num", "transaction_made"]]
df_b_clean = df_b_clean.dropna()
df_b_clean.describe()
```

```python
df_b_clean = df_b_clean.rename(columns={"date_time": "TRANS_DATE", "bill_number": "BILL_NUMBER", "reff_num": "REFF_NUMBER", "transaction_made": "TRANS_AMOUNT"})
df_b_clean.info()
```

```python
df_b_clean
```

```python
# create a new dataframe, with 2 columns, first columns is still date, second column is combination of bill number + ref number + transaction
df_a_combined = pd.DataFrame({'DATE': df_a_clean['TRANS_DATE'], 'DATA': df_a_clean['BILL_NUMBER'].astype(str) + df_a_clean['REFF_NUMBER'].astype(str) + df_a_clean['TRANS_AMOUNT'].astype(str)})
df_a_combined.head()
```

```python
df_b_combined = pd.DataFrame({'DATE': df_b_clean['TRANS_DATE'], 'DATA': df_b_clean['BILL_NUMBER'].astype(str) + df_b_clean['REFF_NUMBER'].astype(str) + df_b_clean['TRANS_AMOUNT'].astype(str)})
df_b_combined.head()
```

```python
merged_df = pd.merge(df_a_combined, df_b_combined, on='DATA', how='inner')
```

```python
new_df = merged_df.rename(columns={"DATE_x": "DATE_A", "DATE_y": "DATE_B"})
# rearrange the columns
new_df = new_df[['DATA', 'DATE_A', 'DATE_B']]
new_df
```

```python
# count NA values
new_df.isna().sum()
```

```python
new_df.info()
```

```python
# change all to string
new_df = new_df.astype(str)
new_df.info()
```

```python
df_complete = pd.DataFrame({'DATA_A': new_df['DATE_A'] + " " + new_df['DATA'], 'DATA_B': new_df['DATE_B'] + " " + new_df['DATA']})
df_complete
```
